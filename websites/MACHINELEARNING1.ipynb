Training Algoritm --Linear Regression

Linear regression is class so we need to make object

```python
from sklearn.linear_model import LinearRegression
```

Making object

```python
lg=LinearRegression()
```

Algorithm training
func for taining =fit()passing training data for training purpose

```python
x_train.isnull().sum()
```

```python
lg.fit(x_train,y_train)
```

Round of function for rounding off

initially we have 2 variable xtrain,ytrain and when we pass it to algorithm we get xtest ,ytest

```python
print("Training Score:",round(lg.score(x_train,y_train)*100 ,2)) #training and testing 
```

```python
print("Testing Score:",round(lg.score(x_test,y_test)*100 ,2))
```

accuracy acheiving  and model evaluation step and it should be more than 90%
diff btw training and testing score it should not be more than five ,if it is then its underfit or overfit
model evaluation == whether model is over fit or under fit or perfect(generalised)
model overfit if training score is high and testing score low 
training score==high and testing score=low==Overfit problem
training score==low and testing score=high/low not assured==underfit problem



Generalised Model 
diff should not more then 5
model shouldnot be overfit and underfit
Accuracy should be high on both train and test data
