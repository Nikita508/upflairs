```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pb
import numpy as np
```

```python
df=sns.load_dataset('tips')
df.head()
```

```python
df.columns=['total_bill', 'tip', 'gender', 'smoker', 'day', 'time', 'member']
```

```python
df
```

```python
# function
#relplot,catlot
# these two functions can create any kind of chart
```

```python
#sns.lineplot()
#sns.boxplot()
#sns.scatterplot()
#sns.countplot()
#sns.barplot()
#sns.histplot()
sns.scatterplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
#relplot=relational plot  used if columns are numerical
# catplot=catgoricalplot   used if one columnn is numerical and other is numerical
```

```python
sns.relplot(x='total_bill',y='tip',data=df)
plt.show()# relatinal plot bydefault give scatter plot
```

```python
sns.relplot(x='total_bill',y='tip',data=df,kind='line') 
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker',col='gender')
plt.show()  # col separated the male and female visitors
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',style='smoker',col='time')
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',data=df,hue='gender',col='day')
plt.show()
```

```python
Female=df[df['gender']=='Female']
Female
```

```python
sns.relplot(x='tip',y='total_bill',data=Female,hue='smoker',col='smoker')
```

```python
sns.relplot(x='tip',y='total_bill',data=Female,hue='smoker',col='day') # row can be used on place of col
```

```python
sns.relplot(x='tip',y='total_bill',data=Female,hue='smoker',row='day')
```

```python
# analysis using single column=univariate analysis
# analysis using two columns=bivariate analysis
# analysis using more then two columns=multivariate analysis
```
