```python
country=['india','china','japan','burma']
population=[900,600,789,543]
explosion=[0,0,0,0.2]
plt.figure(figsize=(8,8))
plt.pie(population,labels=country,autopct='%1.2f%%',wedgeprops={'ec':'k'},explode=explosion,shadow=True)

plt.savefig('my pie plot')
plt.show()
```
